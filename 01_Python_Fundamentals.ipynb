## 1. What is Python?

Python is a **high-level**, **interpreted**, **general-purpose** programming language created by Guido van Rossum in 1991.

- **High-level** → reads almost like English; you don't manage memory by hand.
- **Interpreted** → runs line by line, no separate compile step.
- **General-purpose** → web apps, data science, AI, automation, games… one language for all.

The classic first program in any language:

```python
print("Hello, World!")
```

## 2. Why is Python so popular?

Five reasons that show up again and again:

| Reason | What it means for you |

| **Readable syntax** | Less code, fewer symbols, easier to learn |
| **Huge library ecosystem** | `pip install` almost anything |
| **Beginner friendly** | #1 teaching language worldwide |
| **In-demand jobs** | Data, AI/ML, backend, automation |
| **Massive community** | Answers to almost every question online |

**Readability in action** — compare how little Python needs to add a list of numbers:

```python
# Add the numbers 1..5. Notice how close this is to plain English.
numbers = [1, 2, 3, 4, 5]
total = sum(numbers)
print("The total is", total)
```

In many other languages this same task needs type declarations, a loop, semicolons and braces. Python keeps your attention on the *idea*, not the *ceremony*.

## 4. How does Python run your code?

You write `.py` files (or notebook cells). The **Python interpreter** reads them top to bottom and executes each line immediately — there is no separate compile step like in C++ or Java.

Because of that, Python can be used as an interactive **calculator**:

```python
print(10 + 3)     # addition 13
print(10 - 3)     # subtraction 7
print(10 * 3)     # multiplication 30
print(10 / 3)     # division (always gives a float)
print(10 // 3)    # floor division (whole number)
print(10 % 3)     # remainder (modulus)
print(10 ** 3)    # power (10 cubed) 10*10*10
```

### Comments & the `print()` function

- Anything after `#` on a line is a **comment** — Python ignores it. Comments explain *why*.
- `print()` is how your program talks back to you.

Try editing the strings below and re-running:

```python
# This is a comment - it does nothing when run.
print("Learning")
print("Python")
print("Learning", "Python", "is", "fun", sep=" - ")  # sep changes the separator
```

## 6. Common beginner mistakes

Everyone hits these on day one — spotting them early saves hours:

- Forgetting quotes around text → `print(Hello)` fails, `print("Hello")` works.
- Mismatched quotes → `"hi'` is an error.
- Wrong indentation (Python uses spaces to define blocks).

Here is a broken line and its fix:

```python
# WRONG (uncomment to see the error):
print(Hello)

# RIGHT:
#print("Hello")
```
