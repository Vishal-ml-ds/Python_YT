## 1. What is a Variable?

Think of a variable as a **labelled box** in the computer's memory.
You put a value in, give the box a name, and refer to it by that name later.

In Python you create a variable with a single `=` sign — this is called **assignment**:

```
name = value
```

Python figures out the *type* of the value automatically — you never have to say
`int x = 5` like in Java or C++. This is called **dynamic typing**.

````python
# Assigning values to variables (Python infers the type automatically)
name = "Aarav"          # str  — text
age  = 17               # int  — whole number
price = 299.99          # float — decimal number
is_student = True       # bool — True or False

# Let's print them
print("Name    :", name)
print("Age     :", age)
print("Price   :", price)
print("Student :", is_student)
````

## 2. Naming Rules and Conventions

Python has **hard rules** (break them → error) and **soft conventions** (break them → code still runs, but fellow developers judge you).

### Hard rules
- Only letters, digits, and underscores (`_`).
- **Cannot start with a digit** → `2name` is illegal; `name2` is fine.
- **Cannot be a reserved word** → `for`, `if`, `class`, `True`, `None`, etc.
- Case-sensitive → `Name` and `name` are two different variables.

### Convention — snake_case
Use all lowercase with underscores between words.

| Good ✅ | Bad ❌ |
|---|---|
| `student_name` | `StudentName` (that's for classes) |
| `total_price` | `totalprice` (hard to read) |
| `is_active` | `2fast` (starts with digit → error) |

````python
# Valid names — these all work
student_name = "Diya"
total_price  = 1500
is_active    = True
score2       = 98

print(student_name, total_price, is_active, score2)

# Invalid names — kept as comments so the cell runs without error:
# 2fast = 10        # SyntaxError: cannot start with a digit
# for = 5           # SyntaxError: 'for' is a reserved word
# my-var = 3        # SyntaxError: hyphens not allowed
print("All valid names printed above - no errors!")
````

## 3. Core Data Types

Python has many built-in types. As a beginner you will use these four constantly:

| Type | Python keyword | Example | Real-life |
|---|---|---|---|
| Integer | `int` | `17` | Age, count, rank |
| Float | `float` | `3.14` | Price, temperature, score |
| String | `str` | `"Kabir"` | Name, message, city |
| Boolean | `bool` | `True` | On/off, yes/no, pass/fail |

The built-in function `type()` tells you what type a value is.

````python
# type() reveals the data type — run this and read each output
print(type(42))          # <class 'int'>
print(type(3.14))        # <class 'float'>
print(type("Hello"))     # <class 'str'>
print(type(True))        # <class 'bool'>

# Showing types of your own variables
city     = "Mumbai"
distance = 1350
temp     = 32.5
is_raining = False

print(type(city), type(distance), type(temp), type(is_raining))
````

````python
# A brief mention: complex numbers exist too (for maths/engineering)
z = 3 + 4j
print(type(z))   # <class 'complex'>
print(z.real, z.imag)   # beginners can safely ignore this type for now
````

## 4. Real-Life Example — Student Shopping Profile

Let's model Ananya, a student who is shopping on an e-commerce site.
She has a **name** (str), **age** (int), **wallet balance** (float), and a flag telling us whether she qualifies for a **student discount** (bool).

*(Do this on camera — type each line slowly and explain as you go.)*

````python
# --- Ananya's profile ---
customer_name    = "Ananya"
customer_age     = 19
wallet_balance   = 1500.75
has_student_card = True

# --- Product ---
product_name  = "Python Programming Book"
product_price = 499.00
discount_pct  = 10   # 10 % student discount

# --- Calculation ---
discount_amount = product_price * discount_pct / 100
final_price     = product_price - discount_amount
balance_left    = wallet_balance - final_price

# --- Output ---
print(f"Customer   : {customer_name}, age {customer_age}")
print(f"Student?   : {has_student_card}")
print(f"Product    : {product_name} @ ₹{product_price:.2f}")
print(f"Discount   : ₹{discount_amount:.2f} ({discount_pct}%)")
print(f"You pay    : ₹{final_price:.2f}")
print(f"Wallet left: ₹{balance_left:.2f}")
````

## 5. Type Conversion (Casting)

Sometimes Python stores a number as a string (for example, everything that
comes from `input()` is a string). You need to **convert** it before doing math.

| Function | Does what |
|---|---|
| `int(x)` | Converts x to a whole number |
| `float(x)` | Converts x to a decimal number |
| `str(x)` | Converts x to text |
| `bool(x)` | Converts x to True/False |

**The classic beginner gotcha** — `input()` always returns a string, so `int(input(...))` is the pattern you will type thousands of times.

````python
# int() — string to integer
age_str = "17"        # this looks like a number but Python sees text
age_int = int(age_str)
print(age_str + " is now " + str(age_int + 1))   # str so + works

# float() — string to float
price_str = "299.99"
price_flt = float(price_str)
print(f"Price with tax: {price_flt * 1.18:.2f}")

# str() — number to string
score = 95
print("Your score is " + str(score) + " out of 100")

# bool() — truthiness rules
print(bool(0), bool(1), bool(""), bool("hi"), bool(None))
````

````python
# input() always gives a string — show the pattern with a hardcoded value
# In a real .py script you would write:
#   user_age = int(input("Enter your age: "))

# For this notebook we hardcode the value so nothing pauses:
user_age = int("21")   # same as int(input("Enter your age: "))
print(f"Next year you will be {user_age + 1}")

# WRONG: forgetting int() on input — example (commented to avoid error)
# user_age = "21"            # <- this is a string
# print(user_age + 1)        # TypeError: can only concatenate str to str
````

## 6. Multiple Assignment and the One-Line Swap

Python lets you assign several variables in one line — a handy shortcut
and a great trick to show on camera.

**Multiple assignment on one line:**
```python
a, b, c = 10, 20, 30
```

**Classic swap** (in most languages you need a temp variable; Python doesn't):
```python
a, b = b, a
```

````python
# Multiple assignment in one line
x, y, z = 10, 20, 30
print("x =", x, "  y =", y, "  z =", z)

# Assign the same value to several variables at once
rohan = kabir = diya = 0     # all start with 0 marks
print("Initial marks:", rohan, kabir, diya)

# The famous one-line swap — no temp variable needed!
a = 5
b = 9
print(f"Before swap: a={a}, b={b}")
a, b = b, a
print(f"After  swap: a={a}, b={b}")
````

## 7. Constants — the UPPER_CASE Convention

Python has no built-in `const` keyword. Instead, the community agrees:
> **If a name is ALL_CAPS, treat it as a constant — never reassign it.**

You will see this everywhere in real Python code:

````python
# Constants — by convention, written in UPPER_CASE
PI           = 3.14159
MAX_STUDENTS = 30
BASE_URL     = "https://example.com"
TAX_RATE     = 0.18

print(f"PI = {PI}")
print(f"Max students in batch: {MAX_STUDENTS}")
print(f"Tax rate: {TAX_RATE * 100}%")

# Using a constant in a real calculation
item_cost = 1000
total_with_tax = item_cost * (1 + TAX_RATE)
print(f"Item: ₹{item_cost}  →  Total with GST: ₹{total_with_tax:.2f}")
````

## 8. Common Mistakes

These catch almost every beginner at some point.
Spotting them now will save you hours of head-scratching.

**Mistake 1 — comparing `"10"` (str) with `10` (int)**  
They look the same but Python disagrees.

**Mistake 2 — using a reserved word as a variable name**  
`list = [1, 2, 3]` shadows the built-in `list` type — avoid it.

**Mistake 3 — forgetting `int()` when reading a number from input**  
Adding a string and an integer raises a `TypeError`.

````python
# Mistake 1: "10" (str) vs 10 (int) — they are NOT equal
num_str = "10"
num_int = 10
print(num_str == num_int)       # False — different types!
print(type(num_str), type(num_int))

# Fix: convert before comparing
print(int(num_str) == num_int)  # True — now both are int
````

````python
# Mistake 2: shadowing a built-in name
# list = [1, 2, 3]     # BAD — now you can't use list() as a function!

# Good: use a descriptive name instead
marks_list = [85, 90, 78]
print("Marks:", marks_list)
print("Type :", type(marks_list))
````

````python
# Mistake 3: forgetting int() on input — simulated here
quantity_str = "3"           # imagine this came from input()
unit_price   = 150

# WRONG (commented out — would raise TypeError):
# total = quantity_str + unit_price

# RIGHT: convert first
total = int(quantity_str) * unit_price
print(f"3 items × ₹{unit_price} = ₹{total}")
````
